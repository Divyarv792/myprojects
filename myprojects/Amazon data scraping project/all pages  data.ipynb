```python
import pandas as pd
import os
```

```python
files=[file for file in os.listdir(r'CSV folder')]
allpages_data=pd.DataFrame()
for file in files:
    df=pd.read_csv(r'CSV folder\\'+file)
    allpages_data=pd.concat([allpages_data,df])
    allpages_data.to_csv("all_pages.csv", index=False)
```
